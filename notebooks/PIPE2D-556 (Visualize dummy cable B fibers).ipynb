```python
import numpy as np
from astropy.io import fits

#matplotlib
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 18})
```

```python
# uncomment and load your favorite pfsDetectorMap here
detector_map_fits = fits.open('/Users/nevencaplar/Documents/PFS/Tickets/PIPE2D-547 (Negative values in LSF)/pfsDetectorMap-000000-r1.fits')
```

```python
blue_fibers=[]
for i in [32, 111, 223, 289, 418, 518, 620]:
    vi=list(detector_map_fits[1].data).index(i)
    y_pixel_single_fiber=detector_map_fits[4].data[detector_map_fits[4].data['index']==vi]['knot']
    x_pixel_single_fiber=detector_map_fits[3].data[detector_map_fits[3].data['index']==vi]['value']
    blue_fibers.append([x_pixel_single_fiber,y_pixel_single_fiber[0:len(x_pixel_single_fiber)]])
    
green_fibers=[]
for i in [63, 192, 255, 401, 464, 525, 587]:
    vi=list(detector_map_fits[1].data).index(i)
    y_pixel_single_fiber=detector_map_fits[4].data[detector_map_fits[4].data['index']==vi]['knot']
    x_pixel_single_fiber=detector_map_fits[3].data[detector_map_fits[3].data['index']==vi]['value']
    green_fibers.append([x_pixel_single_fiber,y_pixel_single_fiber[0:len(x_pixel_single_fiber)]])
    
orange_fibers=[]
for i in [12, 60, 110, 161, 210, 259, 341]:
    vi=list(detector_map_fits[1].data).index(i)
    y_pixel_single_fiber=detector_map_fits[4].data[detector_map_fits[4].data['index']==vi]['knot']
    x_pixel_single_fiber=detector_map_fits[3].data[detector_map_fits[3].data['index']==vi]['value']
    orange_fibers.append([x_pixel_single_fiber,y_pixel_single_fiber[0:len(x_pixel_single_fiber)]])
    
yellow_fibers=[]
for i in [347, 400, 449, 466, 545, 593, 641]:
    vi=list(detector_map_fits[1].data).index(i)
    y_pixel_single_fiber=detector_map_fits[4].data[detector_map_fits[4].data['index']==vi]['knot']
    x_pixel_single_fiber=detector_map_fits[3].data[detector_map_fits[3].data['index']==vi]['value']
    yellow_fibers.append([x_pixel_single_fiber,y_pixel_single_fiber[0:len(x_pixel_single_fiber)]])
    
red_fibers=[]
for i in [2, 3, 308, 339, 340, 342, 649,650]:
    vi=list(detector_map_fits[1].data).index(i)
    y_pixel_single_fiber=detector_map_fits[4].data[detector_map_fits[4].data['index']==vi]['knot']
    x_pixel_single_fiber=detector_map_fits[3].data[detector_map_fits[3].data['index']==vi]['value']
    red_fibers.append([x_pixel_single_fiber,y_pixel_single_fiber[0:len(x_pixel_single_fiber)]])
```

```python
plt.figure(figsize=(20,20))
for i in range(len(blue_fibers)):
    plt.plot(blue_fibers[i][0],blue_fibers[i][1],color='blue',lw='3')
    plt.text(x=np.median(blue_fibers[i][0])-100,y=-270,s=np.int(np.median(blue_fibers[i][0])),color='blue',size=16)   
    
for i in range(len(green_fibers)):
    plt.plot(green_fibers[i][0],green_fibers[i][1],color='green',lw='3')
    plt.text(x=np.median(green_fibers[i][0])-100,y=-170,s=np.int(np.median(green_fibers[i][0])),color='green',size=16)   
    
for i in range(len(orange_fibers)):
    plt.plot(orange_fibers[i][0],orange_fibers[i][1],color='orange',lw='3')
    plt.text(x=np.median(orange_fibers[i][0])-100,y=-70,s=np.int(np.median(orange_fibers[i][0])),color='orange',size=16)
    
for i in [1,3,4,5,6]:
    plt.plot(yellow_fibers[i][0],yellow_fibers[i][1],color='gold',lw='3')
    plt.text(x=np.median(yellow_fibers[i][0])-100,y=-370,s=np.int(np.median(yellow_fibers[i][0])),color='gold',rotation='0',size=16)

for i in [0,2]:
    plt.plot(yellow_fibers[i][0],yellow_fibers[i][1],color='gold',lw='3')
    plt.text(x=np.median(yellow_fibers[i][0])-100,y=-470,s=np.int(np.median(yellow_fibers[i][0])),color='gold',rotation='0',size=16)

for i in [1,4,7]:
    plt.plot(red_fibers[i][0],red_fibers[i][1],color='red',lw='3')
    plt.text(x=np.median(red_fibers[i][0])-120,y=-170,s=np.int(np.median(red_fibers[i][0])),color='red',rotation='0',size=16)
 
for i in [0,5,6]:
    plt.plot(red_fibers[i][0],red_fibers[i][1],color='red',lw='3')
    plt.text(x=np.median(red_fibers[i][0])-80,y=-270,s=np.int(np.median(red_fibers[i][0])),color='red',rotation='0',size=16)
    
for i in [3]:
    plt.plot(red_fibers[i][0],red_fibers[i][1],color='red',lw='3')
    plt.text(x=np.median(red_fibers[i][0])-100,y=-370,s=np.int(np.median(red_fibers[i][0])),color='red',rotation='0',size=16)

for i in [2]:
    plt.plot(red_fibers[i][0],red_fibers[i][1],color='red',lw='3')
    plt.text(x=np.median(red_fibers[i][0])-100,y=-470,s=np.int(np.median(red_fibers[i][0])),color='red',rotation='0',size=16)
    
plt.ylim(-500)
plt.title('Avaliable dummy cable B fibers',size=28)
```
